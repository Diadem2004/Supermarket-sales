This dataset contains transaction details of a supermarket, including various attributes such as invoice ID, branch, state, customer type, gender, product line, unit price, quantity, tax, total amount, date, time, payment method, cost of goods sold (COGS), gross margin percentage, gross income, and customer rating. The dataset consists of 1000 rows and 17 columns 


**Data Dictionary**

|       Column Name       | Data Type | Description |
| :---------------------- | :--------------: | :---------- |
| Invoice ID              | object  | Computer generated sales slip invoice identification number |
| Branch                  | object       | Branch of supercenter (3 branches are available identified by A, B and C)           |
| City                    | object       | The City the supermarket is located in         |
| Customer type           | object      | If customer used member card, type is "Members", otherwise "Normal"         |
| Gender                  | object       | Gender of customer           |
| Product line            | object       | General item categories - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel           |
| Unit price              | float64 | Price of each product in $           |
| Quantity                | int64       | Number of products purchased by customer          |
| Tax                     | float64      | 5% tax fee for customer buying          |
| Total                   | float64       | Total price including tax           |
| Date                   | DateTime       | Date of purchase (Record available from January 2019 to March 2019)           |
| Time                    | object       | Purchase time (10am to 9pm)           |
| Payment                 | object       | Payment used by customer for purchase (3 methods are available – Cash, POS, and eNaira)           |
| cogs              | int       | Cost of goods sold           |
| gross margin percentage | float64       | Gross margin percentage           |
| gross income            | float64       | Gross income          |  
| Rating                  | float64       | Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)           |

# Data Visualization

#### QUESTION
* What is the Average Rating?

```python
lines = "--"
color = "red"
labels = "Average rating: {:.2f}"
title = "Rating Distribution"
x = "Ratings"
y = "No of customers"

plt.figure(figsize=(20,7)) #size of graph
plt.hist(Supermarket_Sales['Rating'], bins = 20)# histogram plot with number of bins
plt.title(title)# Graph title
plt.xlabel(x)# X label title
plt.ylabel(y)# y label title
mean_rating = Supermarket_Sales["Rating"].mean()#mean calculation for rating
plt.axvline(mean_rating, color= color, linestyle= lines, label = labels.format(mean_rating))#axis vertical line (axv),axh(axis horizontal line)
plt.legend()
plt.show()
```

#### Findings
* The average rating is 6.97
* The highest number of ratings is 10

#### QUESTION
* Which city has the highest Rating distribution?

```python
import seaborn as sns
#creating a subset of Supermarket sales dataset that contains only Ibadan HQ
Ibadan_HQ_DF=  Supermarket_Sales[Supermarket_Sales['City']=='Ibadan HQ']
#creating a subset of Supermarket sales dataset that contains only Ibadan hub1
Ibadan_hub1_DF= Supermarket_Sales[Supermarket_Sales['City']=='Ibadan hub1']
#creating a subset of Supermarket sales dataset that contains only Ibadan hub2
Ibadan_hub2_DF= Supermarket_Sales[Supermarket_Sales['City']=='Ibadan hub2']


#Plotting a graph that shows the comparison between the three states and their ratings
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,7))
sns.histplot(data= Ibadan_HQ_DF, x='Rating', bins = 15,ax= axes[0], color= "#E65563")
sns.histplot(data=Ibadan_hub1_DF, x='Rating', bins = 15, ax = axes[1], color= 'lightgreen')
sns.histplot(data= Ibadan_hub2_DF, x='Rating', bins= 15, ax = axes[2], color= 'skyblue')
axes[0].set_ylabel('no of customers')
axes[1].set_ylabel('no of customers')
axes[2].set_ylabel('no of customers')
axes[0].set_title('Ibadan HQ Rating Distribution')
axes[1].set_title('Ibadan hub1 Rating Distribution')
axes[2].set_title('Ibadan hub2 Rating Distribution')
mean_rating0= Ibadan_HQ_DF['Rating'].mean()
mean_rating1= Ibadan_hub1_DF['Rating'].mean()
mean_rating2= Ibadan_hub2_DF['Rating'].mean()
axes[0].axvline(mean_rating0, color='red', linestyle= ':', label= 'Average_rating:{:.2f}'.format(mean_rating0))
axes[1].axvline(mean_rating1, color='blue', linestyle= '--', label= 'Average_rating:{:.2f}'.format(mean_rating1))
axes[2].axvline(mean_rating2, color='black', linestyle= '-.', label= 'Average_rating:{:.2f}'.format(mean_rating2))
plt.show()
```

#### FINDINGS
* The city with the highest rating distribution is Ibadan hub1
